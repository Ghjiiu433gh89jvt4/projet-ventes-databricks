Initialisation des catalogues et schémas
À exécuter une seule fois (ou à chaque changement de structure).
S'appuie sur les External Locations déjà créées (adls_bronze_zone, dls_silver_zone, adls_gold_zone) — voir le module ADLS ↔ Unity Catalog.
 `IF NOT EXISTS` rend ce script idempotent : le relancer par erreur ne casse rien.


```python
%sql
CREATE CATALOG IF NOT EXISTS bronze
MANAGED LOCATION 'abfss://bronze@adlsjopadev.dfs.core.windows.net/';

CREATE CATALOG IF NOT EXISTS silver
MANAGED LOCATION 'abfss://silver@adlsjopadev.dfs.core.windows.net/';

CREATE CATALOG IF NOT EXISTS gold
MANAGED LOCATION 'abfss://gold@adlsjopadev.dfs.core.windows.net/';
```

```python
%sql
-- Un schéma "ventes" dans chaque catalogue — c'est le domaine métier de ce
-- projet. Si tu ajoutes un jour un autre domaine (ex: RH, finance), tu
-- créerais un nouveau schéma à côté, pas un nouveau catalogue.

CREATE SCHEMA IF NOT EXISTS bronze.ventes;
CREATE SCHEMA IF NOT EXISTS silver.ventes;
CREATE SCHEMA IF NOT EXISTS gold.ventes;
```

```python
%sql
-- Vérification rapide : la commande suivante doit lister les 3 catalogues
-- et confirmer que le schéma `ventes` existe dans chacun.

SHOW SCHEMAS IN bronze;
```
